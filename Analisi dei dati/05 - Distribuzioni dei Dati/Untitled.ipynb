```python
import pandas as pd
data=pd.read_csv('500_Person_Gender_Height_Weight_Index.csv')
data.info()
```

```python
data.head()
```

```python
#PMF

height_count=data['Height'].value_counts()
print(height_count)
print("Altezza più frequente:",height_count.index[0], "(%d)" % height_count.iloc[0])
print("Altezza meno frequente:",height_count.index[-1], "(%d)" % height_count.iloc[-1])
print(data['Height'].value_counts().sort_index())
```

```python
import matplotlib.pyplot as plt

data['Height'].value_counts().sort_index().plot.bar()
plt.grid()
```

```python
plt.figure(figsize=(16,8))
height_count=data['Height'].value_counts().sort_index()
plt.bar(height_count.index,height_count.values)
plt.xticks(height_count.index,rotation='vertical')
```

```python
## norm histo
                           #normalize=True
data['Height'].value_counts().sort_index()/len(data['Height'])
print("{:0.2f}".format(data['Height'].value_counts(normalize=True).sort_index().sum()))
```

```python
pmf_height_m=data[data['Gender']=='Male']['Height'].value_counts(normalize=True).sort_index()
pmf_height_f=data[data['Gender']=='Female']['Height'].value_counts(normalize=True).sort_index()

plt.figure(figsize=(16,8))
plt.bar(pmf_height_m.index+0.2,pmf_height_m.values,width=1,alpha=0.9)
plt.bar(pmf_height_f.index-0.2,pmf_height_f.values,width=1,alpha=0.7)
plt.legend(['M','F'])
plt.xticks(data['Height'].unique(),rotation='vertical')
plt.grid()
```

```python
## CMF
data['Weight'].value_counts(normalize=True).sort_index().plot.bar()
```

```python
hist=data['Weight'].value_counts(normalize=True).sort_index()
plt.figure(figsize=(16,8))
plt.bar(hist.index,hist.values)
```

```python
pmf_weight_m=data[data['Gender']=='Male']['Weight'].value_counts(normalize=True).sort_index()
pmf_weight_f=data[data['Gender']=='Female']['Weight'].value_counts(normalize=True).sort_index()

plt.figure(figsize=(16,8))
plt.bar(pmf_height_m.index,pmf_height_m.values,width=1,alpha=0.9)
plt.bar(pmf_height_f.index,pmf_height_f.values,width=1,alpha=0.9)
plt.legend((['M','F']))
plt.grid()
plt.show()
```

```python
cdf_weight_m=data[data['Gender']=='Male']['Weight'].value_counts(normalize=True).sort_index().cumsum()
cdf_weight_f=data[data['Gender']=='Female']['Weight'].value_counts(normalize=True).sort_index().cumsum()

plt.figure(figsize=(16,8))
plt.plot(cdf_weight_m.index, cdf_weight_m.values)
plt.plot(cdf_weight_f.index, cdf_weight_f.values)
plt.legend(['M','F'])
plt.grid()
plt.show()







```

```python
titanic = pd.read_csv('https://raw.githubusercontent.com/agconti/kaggle-titanic/master/data/train.csv', index_col='PassengerId')
'''
Pclass Passenger Class (1 = 1st; 2 = 2nd; 3 = 3rd)
survival Survival (0 = No; 1 = Yes)
name Name
sex Sex
age Age
sibsp Number of Siblings/Spouses Aboard
parch Number of Parents/Children Aboard
ticket Ticket Number
fare Passenger Fare (British pound)
cabin Cabin
embarked Port of Embarkation (C = Cherbourg; Q = Queenstown; S = Southampton)
'''
```

```python
titanic.head()
```

```python
a=titanic[titanic['Sex']=='male']['Age'].value_counts(normalize=True).sort_index().cumsum()
b=titanic[titanic['Sex']=='female']['Age'].value_counts(normalize=True).sort_index().cumsum()
plt.figure(figsize=(12,8))
plt.plot(a.index,a.values)
plt.plot(b.index,b.values)
plt.grid()
plt.legend(['M','F'])
plt.show()
```

```python
plt.figure(figsize=(16,8))
val,edges,handler=plt.hist(data['Height'])
plt.xticks(edges)
plt.grid()
plt.show()
print(val,edges,handler)
```

```python
import numpy as np

bins_struges=int(3.3*np.log(len(data['Height'])))
bins_rice=int(2*len(data['Height'])**(1/3))

print(bins_rice,bins_struges)

plt.figure(figsize=(16,6))
plt.subplot(1,2,1)
plt.title('Struger ({} bins)'.format(bins_struges))
plt.hist(data['Height'],bins=bins_struges)
plt.grid()
plt.subplot(1,2,2)
plt.title('Rice ({} bins)'.format(bins_rice))
plt.grid()
plt.hist(data['Height'],bins=bins_rice)
plt.show()
```

```python
plt.figure(figsize=(12,6))
val,edges,_=plt.hist(data['Height'],bins=[140,150,160,170,180,190,200],width=9.6)
plt.xticks(edges)

plt.grid()
plt.show()
print(val,edges)
```

```python
plt.figure(figsize=(12,6))
val,edges,_=plt.hist(data['Height'],bins=[140,150,160,170,180,190,200],width=9.6,density=True)
plt.xticks(edges)

plt.grid()
plt.show()
print(val,edges)
larg=edges[1:]-edges[:-1]
print(larg)
print((larg*val).sum())
```
